```python
# notebook cell: 22.1-2.py

import sys
sys.path.append("..")

from bank.Tower.henry_base import HenryProblem, solve_henry

# Given
T_K = 303.0                     # K (used only to select the correct H in the notebook)
kg_co2_per_kg_h2o = 0.90e-4     # kg CO2 / kg H2O
H_atm = 0.186e4                 # atm (Henry constant at 303 K, appendix value)
MW_CO2 = 0.04401                # kg/mol
MW_H2O = 0.018015               # kg/mol

prob = HenryProblem(
    H=H_atm,
    kg_solute_per_kg_solvent=kg_co2_per_kg_h2o,
    MW_solute=MW_CO2,
    MW_solvent=MW_H2O,
)

res = solve_henry(prob)

P_atm = res["P"]
P_kPa = P_atm * 101.325

print("x_CO2(liq) =", res["x"])
print("P_CO2 (atm) =", P_atm)
print("P_CO2 (kPa) =", P_kPa)
```
